```python
import pandas as pd
import numpy as np
import torch
```

```python
df = pd.read_csv("dataset/hf/hf.csv")
```

```python
perm_cols = ['date', 'PRODUCTREFERENCE']

add_cols = [ 
            # 'ret',
        'aum', 
        'SENT_', 
        # 'PTFSBD', 
        # 'PTFSFX', 
        # 'PTFSCOM', 
        # 'em', 
        # 'sp500', 
        # 'sizespread', 
        # 'bondmkt', 
        # 'creditspread',
        # 'SMB', 'HML','RF', 'mom', 
        # 'con','ipg','tfp','term','def','dei','mkt','lab',
        'confeature', 'tfpfeature', 'ipgfeature', 'termfeature', 'deffeature', 'deifeature', 'mktfeature', 'labfeature',
        'exret',
        ]

# add_cols = ['aum', 'SENT_', 'RF', 'termfeature', 'labfeature', 'exret']
```

```python
event_driven_mix_df = df[df['PRIMARYCATEGORY'] == 'Event Driven']
event_driven_mix_df = event_driven_mix_df.sort_values(by=['date'])
event_driven_mix_df = event_driven_mix_df[perm_cols + add_cols]
event_driven_mix_df = event_driven_mix_df.rename(columns={'PRODUCTREFERENCE': 'partition'})
event_driven_mix_df['date'] = pd.to_datetime(event_driven_mix_df['date'],)
event_driven_mix_df['date'] = event_driven_mix_df['date'] .dt.strftime('%Y-%m-%d %H:%M:%S')
event_driven_mix_df = event_driven_mix_df.dropna()
print(event_driven_mix_df.head(10))
event_driven_mix_df.to_csv('event_driven_mix_10_feature.csv', index=False)
```

```python
all_mix_df = df.sort_values(by=['date'])
all_mix_df = all_mix_df[perm_cols + add_cols]
all_mix_df = all_mix_df.rename(columns={'PRODUCTREFERENCE': 'partition'})
all_mix_df['date'] = pd.to_datetime(all_mix_df['date'],)
all_mix_df['date'] = all_mix_df['date'] .dt.strftime('%Y-%m-%d %H:%M:%S')
all_mix_df = all_mix_df.dropna()
print(all_mix_df.head(10))
all_mix_df.to_csv('all_mix_ex_10_feature.csv', index=False)
```

```python
combined_df = pd.DataFrame(df.date.unique()).rename(columns={0:'date'}).sort_values(by=['date']).reset_index(drop=True)

event_driven_df = df[df['PRIMARYCATEGORY'] == 'Event Driven']
# event_driven_df = df
event_driven_df = event_driven_df[perm_cols + add_cols]

products = event_driven_df.PRODUCTREFERENCE.unique()
print(len(products))

for product in products[:80]:
    # print(len(event_driven_df[event_driven_df.PRODUCTREFERENCE == product]))
    product_df = event_driven_df[event_driven_df.PRODUCTREFERENCE == product]
    product_df = product_df.drop(columns=['PRODUCTREFERENCE'])
    combined_df = combined_df.merge(product_df, on='date', how='left', suffixes=[None, '_' + str(product)]).fillna(0)
    
combined_df
```
